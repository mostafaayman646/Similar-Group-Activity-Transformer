```python
import bz2
import json
import pandas as pd

# 1. Load Event Data (JSON)
with open('FIFA World Cup 2022/Event Data/10510.json', 'rt') as f:
    events_data = json.load(f)

events_df = pd.DataFrame(events_data)
```

```python
# 1. Expand the list of dictionaries into a separate DataFrame
expanded_df = pd.json_normalize(events_df['initialTouch'])

# 2. Drop columns where ALL values are NaN/None
expanded_df = expanded_df.dropna(axis=1, how='all')

# 3. Print the value counts for the remaining valid columns
for col in expanded_df.columns:
    if len(expanded_df[col].value_counts(dropna=True)) > 11:
        continue
    print(f"========== {col} ==========")
    print(expanded_df[col].value_counts(dropna=True))
    print("\n")
```

```python
tracking_frames = []
with bz2.open('FIFA World Cup 2022/Tracking Data/10510.jsonl.bz2', 'rt') as f:
    for line in f:
        if line.strip():
            tracking_frames.append(json.loads(line))

tracking_frames = sorted(tracking_frames, key=lambda x: x['videoTimeMs'])
```

```python
tracking_df = pd.DataFrame(tracking_frames)

# Get a detailed summary (column names, non-null counts, data types)
tracking_df.info()
```
